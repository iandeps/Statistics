## DESCRIPTIVE STATISTICS

```python
# Jaguar vs Panther
resistance = np.array ([
    [10.6, 9.1, 9.3, 9.8, 10.5, 10.4, 9.5, 11, 10.4, 3, 9.8],
    [10.1, 11, 9.1, 20, 9.2, 10.8, 9.9, 9.2, 9.1, 9.1, np.nan]
]).T

print (resistance)
```

```python
# creating DataFrame from numpy array
df = pd.DataFrame(resistance, columns = ['Jaguar', 'Panther'])
df
```

```python
type (df)
```

```python
df['Panther']
```

```python
# Summary of DataFrame
df.info()
```

```python
df.isna()
```

```python
df.dropna()
```

```python
df.fillna(100)
```

```python
df ['Panther'].fillna(df['Panther'].mean())
```

```python
# Summary of statistics
df.describe()
```

```python
# Median
jag_median = df['Jaguar'].median()
jag_median

print(f'Jaguar Median: {jag_median:.2f} ohms')
```

```python
# Median
pan_median = df['Panther'].median()
pan_median

print(f'Jaguar Median: {pan_median:.2f} ohms')
```

```python
# Mode
jag_mode = df['Jaguar'].mode()
pan_mode = df['Panther'].mode()
```

```python
print ('Jaguar Mode:')

for i in jag_mode:
    print (i)
```

```python
print ('Panther Mode:')

for i in pan_mode:
    print (i)
```

```python
# Jaguar IQR

Q1 = df['Jaguar'].quantile(0.25)
Q3 = df['Jaguar'].quantile(0.75)

print(f' Jaguar IQR = {Q3-Q1:.2f} ohms')
```

```python
# Panther IQR

Q1 = df['Panther'].quantile(0.25)
Q3 = df['Panther'].quantile(0.75)

print(f' Panther IQR = {Q3-Q1:.2f} ohms')
```

# BAR CHART <hr style = "border:2.5px solid #37ff00ff"></hr>

```python
jag_mean = df['Jaguar'].mean()
pan_mean = df['Panther'].mean()

plt.figure(figsize=(5,4))
bars = plt.bar(['Jaguar', 'Panther'], [jag_mean,pan_mean], color = "green")
plt.title('Average Resistance by Machine')
plt.ylabel('Resistance (ohms)')
plt.xlabel('Machine')

# Label
plt.bar_label(bars,padding=4)
plt.margins(y=0.5)
plt.show
```
